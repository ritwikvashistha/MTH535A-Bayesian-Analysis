```python
import numpy as np
import matplotlib.pyplot as plt
import json
import os
import seaborn as sns
from scipy.stats import entropy
```

```python
###File structure all the samples of old seeding in old_path ,new seeding in new_path and random seeding in random_path
old_path = "./old/"
new_path = "./new/"
random_path = "./random/"
```

```python
Group_old = open(os.path.join(old_path,"Win.json"),'r')#Change from win -> groupp, Ro_16,quaters,semi,finals
G_old = json.load(Group_old)
Group_old.close()
Group_new = open(os.path.join(new_path,"Win.json"),'r')#
G_new = json.load(Group_new)
Group_new.close()
Group_random = open(os.path.join(random_path,"Win.json"),'r')#
G_random = json.load(Group_random)
Group_random.close()


```

```python
len(G_random)
```

```python
sns.set_theme()
for k,v in G_old.items():
    if k != None:
        prob_new = G_new.get(k,[])
        left = 1000-len(prob_new)
        prob_new.extend([0]*left)
        prob_random = G_random.get(k,[])
        left = 1000-len(prob_random)
        prob_random.extend([0]*left)
    
        ax=sns.displot([v,prob_random,prob_new],kind ='kde',cut=1,legend=False)
        ax.set( ylabel = 'probability_distribution')
        ax.figure.suptitle(k)
        plt.legend(labels=["O","R","N"])
        fig = ax.fig
        fig.savefig(os.path.join("Win_plots",k+".png"))
    
   
```
